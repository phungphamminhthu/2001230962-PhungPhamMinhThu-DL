```python
#11. Tạo ngẫu nhiên 2 tensor kích thước x = (3, 2) và giá trị trong khoảng [5, 10], y = (2, 3) và giá
#trị [15, 20] tương ứng. Hãy giữ giá trị đường chéo của x và thực hiện x + y
x = tf.random.uniform((3, 2), minval=5, maxval=11, dtype=tf.int32)
y = tf.random.uniform((2, 3), minval=15, maxval=21, dtype=tf.int32)
mask = tf.linalg.diag(tf.ones(min(x.shape), dtype=tf.int32))
mask = tf.pad(mask, [[0, x.shape[0]-mask.shape[0]], [0, x.shape[1]-mask.shape[1]]])
x_diag = x * mask
result = tf.matmul(x_diag, y)
print("x:\n", x)
print("x (giữ đường chéo):\n", x_diag)
print("y:\n", y)
print("Kết quả x @ y:\n", result)
```

```python
#12. Viết hàm tạo 2 tensor ngẫu nhiên có kích thước x = (3, 2, 2), y = (3, 2, 2). Viết hàm tính cộng,
#trừ, nhân, chia sau đó chuyển các hàm này vào trong graph.
def create_tensors():
    x = tf.random.uniform((3, 2, 2))
    y = tf.random.uniform((3, 2, 2))
    return x, y
@tf.function
def tensor_ops(x, y):
    add = x + y
    sub = x - y
    mul = x * y
    div = x / y
    return add, sub, mul, div
x, y = create_tensors()
add, sub, mul, div = tensor_ops(x, y)
print("x:\n", x)
print("y:\n", y)
print("x + y:\n", add)
print("x - y:\n", sub)
print("x * y:\n", mul)
print("x / y:\n", div)
```

```python
#13. Xây dựng model để tính tổng hai tensor x, y
from tensorflow.keras import Input, Model
from tensorflow.keras.layers import Add
input_x = Input(shape=(2,))
input_y = Input(shape=(2,))
output = Add()([input_x, input_y])
model = Model(inputs=[input_x, input_y], outputs=output)
x = tf.constant([[1.0, 2.0]])
y = tf.constant([[3.0, 4.0]])

result = model([x, y])
print(result)
```

```python
#14. Xây dựng model để tính hiệu và tích hai tensor x, y
from tensorflow.keras.layers import Subtract, Multiply
input_x = Input(shape=(2,))
input_y = Input(shape=(2,))
sub = Subtract()([input_x, input_y])
mul = Multiply()([input_x, input_y])
model = Model(inputs=[input_x, input_y], outputs=[sub, mul])
x = tf.constant([[5.0, 6.0]])
y = tf.constant([[2.0, 3.0]])

result_sub, result_mul = model([x, y])

print("Hiệu:", result_sub)
print("Tích:", result_mul)
```

```python
#15. Xây dựng model để tính chu vi hình chữ nhật P = (a +b) * 2
from tensorflow.keras.layers import Lambda
a = Input(shape=(1,))
b = Input(shape=(1,))
sum_ab = Add()([a, b])

P = Lambda(lambda x: x * 2)(sum_ab)

model = Model(inputs=[a, b], outputs=P)
a_val = tf.constant([[3.0]])
b_val = tf.constant([[4.0]])

result = model([a_val, b_val])
print("Chu vi:", result)
```

```python
#16. Xây dựng model để tính diện tích toàn phần hình hộp chữ nhật
#Sxung quanh = (a + b) * c * 2
#Stoàn phần = (a * b) * 2 + Sxung quanh
from tensorflow.keras.layers import Multiply
a = Input(shape=(1,))
b = Input(shape=(1,))
c = Input(shape=(1,))

sum_ab = Add()([a, b])

sxq = Multiply()([sum_ab, c])
sxq = Lambda(lambda x: x * 2)(sxq)
ab = Multiply()([a, b])
two_ab = Lambda(lambda x: x * 2)(ab)

stp = Add()([two_ab, sxq])

model = Model(inputs=[a, b, c], outputs=stp)

# Test
a_val = tf.constant([[2.0]])
b_val = tf.constant([[3.0]])
c_val = tf.constant([[4.0]])

result = model([a_val, b_val, c_val])
print("Diện tích toàn phần:", result)
```

```python
#17. Xây dựng model chứng minh hai cạnh bên hình thang song song theo điều kiện
import tensorflow as tf
from tensorflow.keras import Input, Model
from tensorflow.keras.layers import Lambda

AD = Input(shape=(1,))
BC = Input(shape=(1,))
AB = Input(shape=(1,))
CD = Input(shape=(1,))

# So sánh
cond1 = Lambda(lambda x: tf.abs(x[0] - x[1]) < 1e-5,
               output_shape=(1,))([AD, BC])

cond2 = Lambda(lambda x: tf.abs(x[0] - x[1]) < 1e-5,
               output_shape=(1,))([AB, CD])

# Logical AND (phải chỉ rõ shape)
parallel = Lambda(lambda x: tf.logical_and(x[0], x[1]),
                  output_shape=(1,))([cond1, cond2])

model = Model(inputs=[AD, BC, AB, CD], outputs=parallel)
import tensorflow as tf

AD_val = tf.constant([[3.0]])
BC_val = tf.constant([[3.0]])
AB_val = tf.constant([[5.0]])
CD_val = tf.constant([[5.0]])

result = model([AD_val, BC_val, AB_val, CD_val])
print(result)
```

```python
#18. Xây dựng model chứng minh hai cạnh bên hình thang bằng nhau theo điều kiện
import tensorflow as tf
from tensorflow.keras import Input, Model
from tensorflow.keras.layers import Lambda

AD = Input(shape=(1,))
BC = Input(shape=(1,))
AB = Input(shape=(1,))
CD = Input(shape=(1,))

cond1 = Lambda(lambda x: tf.abs(x[0] - x[1]) < 1e-5,
               output_shape=(1,))([AD, BC])

cond2 = Lambda(lambda x: tf.abs(x[0] - x[1]) < 1e-5,
               output_shape=(1,))([AB, CD])

parallel = Lambda(lambda x: tf.logical_and(x[0], x[1]),
                  output_shape=(1,))([cond1, cond2])

model = Model(inputs=[AD, BC, AB, CD], outputs=parallel)
AB_val = tf.constant([[5.0]])
CD_val = tf.constant([[5.0]])
AD_val = tf.constant([[3.0]])
BC_val = tf.constant([[3.0]])

print(model([AB_val, CD_val, AD_val, BC_val]))
```

```python
#19. Xây dựng model chứng minh NA = NC theo mô tả sau với AM, MB, AN, NC là đầu vào
import tensorflow as tf
from tensorflow.keras import Input, Model
from tensorflow.keras.layers import Lambda

AM = Input(shape=(1,))
MB = Input(shape=(1,))
AN = Input(shape=(1,))
NC = Input(shape=(1,))
cond1 = Lambda(
    lambda x: tf.less(tf.abs(x[0] - x[1]), 1e-5),
    output_shape=(1,)
)([AM, MB])

cond2 = Lambda(
    lambda x: tf.less(tf.abs(x[0] - x[1]), 1e-5),
    output_shape=(1,)
)([AN, NC])

result = Lambda(
    lambda x: tf.logical_and(x[0], x[1]),
    output_shape=(1,)
)([cond1, cond2])

model = Model(inputs=[AM, MB, AN, NC], outputs=result)

# Test
AM_val = tf.constant([[2.0]])
MB_val = tf.constant([[2.0]])
AN_val = tf.constant([[3.0]])
NC_val = tf.constant([[3.0]])

print("Kết quả NA = NC:", model([AM_val, MB_val, AN_val, NC_val]))
```
