```python
#1.Tạo mảng 20 số nguyên ngẫu nhiên kích thước 3 x 3 có giá trị trong khoảng [0, 20]. Viết
#chương trình kiểm tra tất cả các phần tử trong mảng khác 0 (tìm hiểu hàm all của numpy)
import numpy as np

# Tạo ma trận 3x3 với số nguyên ngẫu nhiên từ 0 đến 20
arr = np.random.randint(0, 21, size=(3, 3))

print("Ma trận:")
print(arr)

# Kiểm tra tất cả phần tử khác 0
check = np.all(arr != 0)

print("Tất cả phần tử khác 0:", check)
```

```python
#19. Viết chương trình Pandas tìm những phần tử trong Series x mà không có trong Series y (Tìm
#hiểu hàm isin của pandas)
a = pd.Series([1, 2, 3, 4, 5])
b = pd.Series([2, 4, 6, 8, 10])
print('Series a:\n', a)
print('\nSeries a:\n', b)
kq = a[~a.isin(b)]
print('\nKết quả các phần tử của a không có trong b:\n', kq)
```

```python
#20. Tìm hiểu phép toán numpy.union1d, numpy.intersect1d (hội, giao) và viết chương trình minh
#họa 2 hàm này trên 2 Series x, y ngẫu nhiên có 20 phần tử
x = pd.Series(np.random.randint(0, 30, 20))
y = pd.Series(np.random.randint(0,30, 20))
print('Series x:\n', x)
print('\nSeries y:\n', y)
union_res = np.union1d(x, y)
print('\nPhép hội:"\n', union_res)
intersection_res = np.intersect1d(x, y)
print('\nPhép giao:\n', intersection_res)
```

```python
#21. Viết chương trình Pandas thêm series y vào series x theo chiều dọc (vertical), chiều ngang
#(horizontal) (Tìm hiểu hàm append và concat của pandas)
# Tạo 2 Series
x = pd.Series([1, 2, 3, 4, 5])
y = pd.Series([6, 7, 8, 9, 10])

print("Series x:\n", x)
print("\nSeries y:\n", y)

vertical_concat = pd.concat([x, y], ignore_index=True)
print("\nGhép theo chiều dọc (concat):\n", vertical_concat)

# --- Ghép theo chiều ngang (horizontal) ---
horizontal_concat = pd.concat([x, y], axis=1)
print("\nGhép theo chiều ngang (concat):\n", horizontal_concat)
```
